```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 4000 


tables = pd.read_html("https://www.jhowell.net/cf/scores/Missouri.htm")

# print(tables[1])
#drop the first one (some random stuff)
tables.pop(0)
# print(tables[0])



scores = []
print(tables[0].iloc[1][5])
for season in range(len(tables)):
    # we have all the historic scores
    # now lets just get every score
    # since score is in the 4/5 column, we can just get that
    
    # but we have to make this a two loop thing
    # one for each season, and then another for game in that season
    
    for game in range(len(tables[season])):
        # this gets us the scores of mizzou and the opposing team
        magic_line = (str(tables[season].iloc[game][4]) + " - " + str(tables[season].iloc[game][5]))
        scores.append(magic_line)

    
    
print(scores)
print(len(scores))

#nice, we got the scores, lets clean them up a bit
```

```python
# clean up the confrence name
scores = [score for score in scores if not 
          any(conference in score for conference in [
            "(SEC)", "(Big 12)", "(Big 8)", "(Big 7)", "(Big 6)", "(MVIAA)", "(Independent)"])]

print(len(scores))

# right, now we have an issue where the last entry of every season is the combined score
print(scores)
```
